# Python - Data Toolkit (Assignment)
## 1. What is NumPy, and why is it widely used in Python?
### NumPy (Numerical Python) is a library for fast numerical operations.
### It is widely used because it supports multi-dimensional arrays, mathematical functions, and vectorized operations.

## 2. How does broadcasting work in NumPy?
### Broadcasting allows NumPy to perform operations on arrays of different shapes
### without explicitly copying data. Example: adding a scalar to a vector.

## 3. What is a Pandas DataFrame?
### A DataFrame is a 2D labeled data structure in Pandas, similar to a spreadsheet.
### It has rows (index) and columns.

## 4. Explain the use of the groupby() method in Pandas
### groupby() is used to split data into groups, apply functions, and combine results.
### Example: finding average sales per region.

## 5. Why is Seaborn preferred for statistical visualizations?
### Seaborn is built on top of Matplotlib and provides simple functions
### for beautiful, statistical plots like histograms, boxplots, heatmaps, etc.

## 6. What are the differences between NumPy arrays and Python lists?
### NumPy arrays are faster, consume less memory, and support vectorized operations.
### Python lists are more flexible but slower for numerical calculations.

## 7. What is a heatmap, and when should it be used?
### A heatmap is a colored 2D representation of data.
### It is useful for showing correlations or intensity (e.g., correlation matrix).

## 8. What does the term “vectorized operation” mean in NumPy?
### Vectorization means applying operations directly on arrays without loops.
### Example: arr * 2 (applies to all elements at once).

## 9. How does Matplotlib differ from Plotly?
### Matplotlib is static (great for basic plots).
### Plotly is interactive (zoom, hover, dynamic dashboards).

## 10. What is the significance of hierarchical indexing in Pandas?
### Hierarchical indexing (MultiIndex) allows multiple levels of row/column labels.
### Useful for complex datasets (e.g., sales data by country & year).

## 11. What is the role of Seaborn’s pairplot() function?
### pairplot() shows pairwise relationships between multiple numeric columns
### using scatterplots and histograms.

## 12. What is the purpose of the describe() function in Pandas?
### describe() gives summary statistics of a DataFrame (mean, std, min, max, quartiles).

## 13. Why is handling missing data important in Pandas?
### Missing data can lead to errors or wrong analysis.
### Pandas provides dropna() and fillna() to handle missing values.

## 14. What are the benefits of using Plotly for data visualization?
### Plotly provides interactive, web-based, and highly customizable plots.
# Great for dashboards and presentations.

## 15. How does NumPy handle multidimensional arrays?
### NumPy uses n-dimensional arrays (ndarray).
# You can create 1D, 2D, 3D, or higher dimensional arrays for scientific computing.

## 16. What is the role of Bokeh in data visualization?
### Bokeh is used for interactive visualizations in the browser.
### It is great for dashboards and handling large streaming datasets.

## 17. Explain the difference between apply() and map() in Pandas
### map() works on Series (single column).
### apply() works on both DataFrame and Series, allowing more flexibility.

## 18. What are some advanced features of NumPy?
### Linear algebra, Fourier transforms, random number generation, broadcasting.

## 19. How does Pandas simplify time series analysis?
### Pandas supports date parsing, resampling, shifting, rolling windows, etc.
### Makes handling time-based data easier.

## 20. What is the role of a pivot table in Pandas?
### A pivot_table summarizes data by grouping and aggregating it.
### Example: average sales per region per year.

## 21. Why is NumPy’s array slicing faster than Python’s list slicing?
### NumPy arrays are stored in continuous memory blocks (C language backend).
### Python lists are stored as objects with more overhead.

## 22. What are some common use cases for Seaborn?
### Correlation analysis (heatmaps), distribution analysis (histplot, distplot),
### categorical data analysis (boxplot, barplot), regression plots, etc.


```python
# Q1. How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Array:\n", arr)
print("Row sums:", np.sum(arr, axis=1))
```

```python
# Q2. Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd
data = {"Name": ["A", "B", "C"], "Age": [20, 25, 30]}
df = pd.DataFrame(data)
print("Mean Age:", df["Age"].mean())
```

```python
# Q3. Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 7, 11]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python

# Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
data = pd.DataFrame({"A": [1, 2, 3], "B": [3, 2, 1], "C": [2, 3, 4]})
corr = data.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python

# Q5. Generate a bar plot using Plotly
import plotly.express as px
data = {"Fruit": ["Apple", "Banana", "Orange"], "Quantity": [10, 15, 7]}
fig = px.bar(data, x="Fruit", y="Quantity", title="Q5 - Bar Plot")
fig.show()
```

```python
# Q6. Create a DataFrame and add a new column based on an existing column
df = pd.DataFrame({"Name": ["A", "B", "C"], "Marks": [50, 65, 80]})
df["Result"] = ["Pass" if m >= 60 else "Fail" for m in df["Marks"]]
print("\nQ6 - DataFrame:\n", df)
```

```python
# Q7. Write a program to perform element-wise multiplication of two NumPy arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("\nQ7 - Element-wise multiplication:", a * b)
```

```python
# Q8. Create a line plot with multiple lines using Matplotlib
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.title("Q8 - Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Q9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
df = pd.DataFrame({"Name": ["A", "B", "C"], "Score": [40, 75, 90]})
filtered = df[df["Score"] > 50]
print("\nQ9 - Filtered DataFrame:\n", filtered)
```

```python

# Q10. Create a histogram using Seaborn to visualize a distribution
data = [1, 2, 2, 3, 3, 3, 4, 5, 5, 5]
sns.histplot(data, bins=5, kde=True)
plt.title("Q10 - Histogram")
plt.show()
```

```python
# Q11. Perform matrix multiplication using NumPy
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("\nQ11 - Matrix Multiplication:\n", np.matmul(A, B))
```
